# Menganalisis performa penjualan e-commerce

```python
orders_df.groupby(orders_df['order_status']).count()
```

Untuk menganalisis bagaimana performa penjualan e-commerce, analisis dilakukan dengan melihat total order yang berhasil dideliver pada customer melalui kolom 'order_status'. Kolom tersebut berisi status order customer. Order status 'delivered' menyatakan bahwa transaksi berhasil dan barang telah dikirim.  Karena pada kolom order status masih terdapat beberapa data yang menyatakan order tidak berhasil seperti status 'cancelled' atau 'unavailable', maka kita dapat melakukan analisis penjualan untuk order status 'delivered'.

### Analisis penjualan berdasarkan tahun

```python
# mengambil data dimana order telah berhasil diterima oleh customer
sales_df = orders_df[orders_df['order_status']=='delivered']
sales_df['year']=sales_df['order_purchase_timestamp'].dt.year
grouped_sales_by_year = sales_df.groupby(sales_df['year']).count().reset_index()
grouped_sales_by_year
```

```python
plt.figure(figsize=(8,6))
plt.plot(grouped_sales_by_year['year'],grouped_sales_by_year['order_purchase_timestamp'],marker='o')
plt.xticks(grouped_sales_by_year['year'])
plt.yticks(grouped_sales_by_year['order_purchase_timestamp'])
plt.xlabel('year')
plt.ylabel('orders delivered')
plt.title('orders delivered by year')
plt.show()
```

Dapat dilihat dari data di atas, grafik order delivered memiliki tren menaik dari tahun 2016 hingga tahun 2018. Hal ini menyatakan penjualan e-commerce mengalami kenaikan dari tahun 2016 hingga 2018. 

### Analisis penjualan berdasarkan bulan

```python
sales_df['month']=sales_df['order_purchase_timestamp'].dt.month
sales_df=sales_df.astype({'month':'int64'})

sales_df_2016 = sales_df[sales_df['year']==2016]
sales_df_2017 = sales_df[sales_df['year']==2017]
sales_df_2018 = sales_df[sales_df['year']==2018]
```

```python
sales_df
```

```python
grouped_orders_by_month_2016 = sales_df_2016.groupby(sales_df_2016['month']).count().reset_index()

months=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
plt.figure(figsize=(8,6))
month=[]
for i in grouped_orders_by_month_2016['month']:
    month.append(months[i-1])
plt.plot(month,grouped_orders_by_month_2016['order_purchase_timestamp'],marker='o')
plt.xticks(month)
plt.yticks(grouped_orders_by_month_2016['order_purchase_timestamp'])
plt.ylabel('orders delivered')
plt.title('orders delivered by month in 2016')
plt.show()
```

```python
grouped_orders_by_month_2016
```

Berdasarkan data, pada tahun 2016 hanya tercatat 267 penjualan, yaitu pada bulan september sebanyak 1, pada bulan oktober sebanyak 265, dan bulan desember sebanyak 1.

```python
grouped_orders_by_month_2017 = sales_df_2017.groupby(sales_df_2017['month']).count().reset_index()

plt.figure(figsize=(15,12))
month=[]
for i in grouped_orders_by_month_2017['month']:
    month.append(months[i-1])
plt.plot(month,grouped_orders_by_month_2017['order_purchase_timestamp'],marker='o')
plt.xticks(month)
plt.yticks(grouped_orders_by_month_2017['order_purchase_timestamp'])
plt.ylabel('orders delivered')
plt.title('orders delivered by month in 2017')
plt.show()
```

# Menganalisis produk yang paling banyak diorder

```python
order_items_df
```

Untuk menganalisis produk apa yang paling banyak diorder, kita dapat menganalisis order_items dataset.

```python
grouped_ordered_products = order_items_df.groupby(order_items_df['product_id']).count().reset_index().sort_values(by='order_id',ascending=False)
grouped_ordered_products.head(10)
```

```python
# menambah kolom category name berdasarkan product id
grouped_ordered_products = pd.merge(grouped_ordered_products,products_df[['product_id','product_category_name']],on = 'product_id')
```

```python
grouped_ordered_products.info()
```

Selanjutnya, kita dapat mengubah nama produk menjadi bahasa inggris berdasarkan dataset product_category_name_translation.

```python
product_category_name_translation.info()
```

```python
product_category_name_english = []
for name in grouped_ordered_products['product_category_name']:
    add=True
    for n,en_name in zip(product_category_name_translation['product_category_name'],product_category_name_translation['product_category_name_english']):
        if name == n:
            product_category_name_english.append(en_name)
            add = False
    if add:
         product_category_name_english.append(n)
```

```python
grouped_ordered_products['product_category_name_english']=product_category_name_english
grouped_ordered_products
```

```python
# mengurutkan produk yang paling banyak diorder
top_ordered_products = grouped_ordered_products.groupby(grouped_ordered_products['product_category_name_english']).agg({
    'order_id':'sum'
}).reset_index()
top_ordered_products = top_ordered_products.sort_values(by='order_id',ascending=False)
```

```python
top_ordered_products
```

```python
top10_products = top_ordered_products.head(10)
colors = ["#00ffaa", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3","#D3D3D3","#D3D3D3","#D3D3D3","#D3D3D3","#D3D3D3"]
plt.figure(figsize=(12,10))
plt.barh(top10_products['product_category_name_english'],top10_products['order_id'])
plt.title('Top 10 most ordered products')

sns.barplot(x='order_id', y='product_category_name_english', data=top10_products, palette=colors)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis ='y', labelsize=12)
plt.show()
```

Berdasarkan data, dapat dilihat produk bed_bath_table merupakan produk yang paling banyak diorder dari e-commerce dengan total order mencapai 11115 order.

# RFM analysis

```python
orders_df.sort_values(by='order_purchase_timestamp',ascending=False)
```

```python
order_payments_df.groupby(order_payments_df['order_id']).count().sort_values(by='payment_value',ascending=False)
```

```python
order_payments_df[order_payments_df['order_id']=='bfbd0f9bdef84302105ad712db648a6c']
```

```python
# menggabungkan kolom payment value dengan orders dataset berdasarkan order_id
rfm_orders_df = orders_df
grouped_payments_by_id = order_payments_df.groupby(order_payments_df['order_id']).sum().reset_index()
rfm_orders_df = rfm_orders_df.merge(grouped_payments_by_id[['order_id','payment_value']],on='order_id',how='left')
rfm_orders_df.info()
```

```python
# mencari order_id dengan kolom payment_value kosong
rfm_orders_df[rfm_orders_df['payment_value'].isna()]
```

```python
rfm_df = rfm_orders_df.groupby(by="customer_id", as_index=False).agg({
    "order_purchase_timestamp": "max", 
    "order_id": "nunique", 
    "payment_value": "sum" 
})
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]
 
# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = orders_df["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)
 
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
```

```python
rfm_orders_df.sort_values(by='order_purchase_timestamp',ascending=False)
```

```python
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(27, 30))
colors = ["#00ffaa"]*5
 
sns.barplot(y="recency", x="customer_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)
 
sns.barplot(y="frequency", x="customer_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
 
sns.barplot(y="monetary", x="customer_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)
 
plt.suptitle("Best Customer Based on RFM Parameters (customer_id)", fontsize=30)
plt.show()
```

Berdasarkan hasil analisis segmentasi customer berdasarkan RFM, frekuensi order customer pada e-commerce paling banyak hanya 1 kali. Berdasarkan monetary terdapat beberapa customer yang melakukan transaksi yang cukup besar dalam satu transaksi dengan transaksi tertinggi hampir mencapai 14000 dalam satu transaksi. Berdasarkan recency interaksi customer dengan e-commerce cukup bervariasi dengan interaksi terbaru berdasarkan dataset adalah 2018-10-17 oleh customer dengan id pertama pada grafik recency.
